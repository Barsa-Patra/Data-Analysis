```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("latest Covid-19 India Status1.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include=object)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.barplot("State/UTs","Total Cases",data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot("State/UTs","Active",data=df,palette=["DarkSeaGreen","lime","palegreen","yellowgreen","green","darkgreen"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot("State/UTs","Deaths",data=df,palette=["red","orange","darkred","Tomato","RosyBrown","brown"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State/UTs'],df['Active Ratio (%)'],color='Purple',marker='o')
plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)
plt.xlabel('State/UTs', fontsize=15)

plt.ylabel('Active Ration', fontsize=15)
plt.title("Line Plot for Active Ratio",fontsize=15)
plt.show()

```

```python
plt.figure(figsize=(8,8))
labels = df.index
plt.pie(x='Active', data=df.head(), labels='State/UTs',startangle=90,autopct='%.1f%%')
plt.title("Active Cases in Top 5 states in India", fontsize = 24) 
plt.tight_layout() 
plt.show()

```
